```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


#Leyendo el archivo CSV
data= pd.read_csv("fetal_health.csv", sep=",", header=[0])
print(data.describe())
X = data.iloc[:, :-1].values
print(X[0])
Y = data.iloc[:,21].values
print(Y)

from sklearn.impute import SimpleImputer
from sklearn import preprocessing

#Preprosesamiento 1, rellenar con la media los espacios faltantes

imputer = SimpleImputer(missing_values=np.NaN, strategy='mean')
imputer = imputer.fit(X)
X= imputer.transform(X)

#Preprosesamiento 2, normalizar los datos

X=preprocessing.normalize(X)
#print(X)

from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
#clasificador=MLPClassifier(hidden_layer_sizes=(128,64,32), solver='adam',max_iter=500)
#clasificador=MLPClassifier()
clasificador = DecisionTreeClassifier()
#clasificador =LogisticRegression(random_state = 0)
clasificador.fit(X,Y)
y_obtenido=clasificador.predict(X)
matriz=confusion_matrix(y_obtenido,Y)

print("==================================================")
print("Matriz de confusi贸n :                             | ")
print("==================================================")
print(matriz)

print("==================================================")
print("Precisi贸n           :                            | ")
print("==================================================")
print(precision_score(y_obtenido,Y,average='micro'))
print("==================================================")

print("==================================================")
print("Splits             :                            | ")
print("==================================================")
```

```python
shufflesplit=StratifiedShuffleSplit(n_splits=100, test_size=0.20, random_state=0)
cv=shufflesplit.split(X,Y)
confiabilidad1=np.zeros(100)
confiabilidad2=np.zeros(100)
for indice, (train_index, test_index) in enumerate( cv):
    print("-----")
    print("Split: ",indice)
    clasificador.fit(X[train_index],Y[train_index])
    y_obtenido = clasificador.predict(X[test_index])
    print(Y[test_index])
    print(y_obtenido)
    print("")
    print("Matriz de Confusi贸n")
    print(confusion_matrix(y_obtenido,Y[test_index]))
    print("Precisi贸n")
    c1=precision_score(y_obtenido, Y[test_index], average='micro')
    print(c1)
    confiabilidad1[indice]=c1
   
```
